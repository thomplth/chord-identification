### Note

```python
from music21 import note
```

```python
n1 = note.Note()
n1.show()
```

```python
n1.octave = 5
n1.show()
```

```python
# transpose
n2 = note.Note('D')
n3 = n2.transpose('M3') # Major 3

print('n2:\t', n2.name, '\t', n2.pitch.accidental)
print('n3:\t', n3.name, '\t', n3.pitch.accidental)

n2.transpose('M3', inPlace=True)
print('\nn2:\t', n2.name, '\t', n2.pitch.accidental)
```

```python
print('note fullName: \t', n1.fullName)
print('note isChord: \t', n1.isChord)
print('note isNote: \t', n1.isNote)
print('note isRest: \t', n1.isRest)
print('note isStream: \t', n1.isStream)
print('note name: \t', n1.name)
print('note nameWithOctave: \t', n1.nameWithOctave)
print('note octave: \t', n1.octave)
print('note offset: \t', n1.offset)
print('note pitches: \t', n1.pitches)
print('note quarterLength: \t', n1.quarterLength)
print('note step: \t', n1.step)
```

```python
r = note.Rest()
r.show()
```

```python
# duration
print(n1.duration.quarterLength)
n1.show()

n1.duration.quarterLength = 4
print(n1.duration.quarterLength)
n1.show()
```

```python
from music21 import pitch
```

### Stream

```python
from music21 import stream
```

```python
s1 = stream.Stream()
s1.show()
```

```python
s1.append(n1)
s1.append(n2)
s1.append(n3)
s1.repeatAppend(n4, 4)
s1.show()
```

```python
for n in s1:
    print(n.pitch.unicodeName, n.offset)
```
